```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import keras
from keras import models, layers
import tensorflow as tf
```

```python
df = pd.read_csv('./stars.csv')
df
```

```python
le_color = LabelEncoder()
df['Star color'] = le_color.fit_transform(df['Star color'])
le_class = LabelEncoder()
df['Spectral Class'] = le_class.fit_transform(df['Spectral Class'])
df
```

```python
x = df.drop(['Spectral Class'], axis=1)
y = df['Spectral Class']
```

```python
y.value_counts()
```

```python
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
# y_scaled = sc.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2)
```

```python
model = models.Sequential([
    layers.Dense(128, input_shape=(6,)),
    layers.Activation('relu'),
    layers.Dense(128),
    layers.Activation('relu'),
    layers.Dense(7),
    layers.Activation('softmax')
])
model.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(), optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=32, batch_size=8, validation_data=(x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test)
```
