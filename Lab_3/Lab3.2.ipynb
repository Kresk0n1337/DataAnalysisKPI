```python
train = image_dataset_from_directory('./Animal Classification/Animal Classification/Training Data', label_mode = 'categorical', image_size=(256,256))
test = image_dataset_from_directory('./Animal Classification/Animal Classification/Testing Data', label_mode = 'categorical', image_size=(256,256))
val = image_dataset_from_directory('./Animal Classification/Animal Classification/Validation Data', label_mode = 'categorical', image_size=(256,256))
```

```python
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(256,256, 3)),
    layers.Conv2D(64, 3, padding='same'),
    layers.Activation('relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(128, 3, padding='same'),
    layers.Activation('relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(256, 3, padding='same'),
    layers.Activation('relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128),
    layers.Activation('relu'),
    layers.Dense(5),
    layers.Activation('softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(train, epochs=64, batch_size=32, validation_data=val)
```

```python
model.evaluate(test)
```
